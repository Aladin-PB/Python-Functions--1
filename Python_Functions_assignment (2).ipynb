4. What is the purpose of the `return` statement in a Python function?
- Ans. The return statement in a Python function serves two primary purposes:
- Exiting the function:   
When a return statement is encountered, the function's execution is immediately terminated, and control is passed back to the caller. Any code within the function after the return statement is not executed.
- Returning a value:    
The return statement can optionally return a value to the caller. This value can be of any data type, including numbers, strings, lists, tuples, dictionaries, or even other functions. If no value is specified, the function returns None by default.

7. What are the advantages of using generators over regular functions?
- Ans : Generators offer several advantages over regular functions, primarily related to memory efficiency, lazy evaluation, and the ability to handle infinite sequences.-
- Memory Efficiency:
Generators produce values on demand using the yield keyword, and do not store the entire sequence in memory. This is especially beneficial when dealing with large datasets or sequences, as it avoids memory overflow issues. Regular functions, on the other hand, generate all values at once and store them in memory before returning.
- Lazy Evaluation:
Generators compute values only when they are needed. This "lazy" approach can save computational resources, especially when not all generated values are required. Regular functions compute all values upfront, regardless of whether they are used.
- Handling Infinite Sequences:
Generators can represent infinite sequences. Because they produce values one at a time, they can theoretically continue indefinitely, as long as there is sufficient memory. Regular functions cannot handle infinite sequences because they must produce all values before returning.
- Cooperative Multitasking:
Generators can pause their execution using the yield keyword and resume later from where they left off. This allows for a form of cooperative multitasking, where a generator can relinquish control to other parts of the program and then resume its operation later. Regular functions run to completion in one go.
- Resource Pipelining:
Generators facilitate resource pipelining, where data processing tasks are divided into a sequence of generator functions. Each generator processes a segment of data and passes it to the next, creating an efficient processing pipeline that minimizes memory usage.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- Ans.  The map(), filter() and reduce() functions bring a bit of functional programming to Python. All three of these are convenience functions that can be replaced with List Comprehensions or loops, but provide a more elegant and short-hand approach to some problems.
- map():
Applies a given function to each item in an iterable and returns an iterator that yields the results. It transforms each element individually.
- filter():
Creates a new iterator containing only the items from the original iterable for which a given function returns True. It selects elements based on a condition.
- reduce():
Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It aggregates elements.
-Ex.
     
     
     '''map(): squares each number'''
     numbers = [1, 2, 3, 4, 5]
     squared_numbers = map(lambda x: x**2, numbers)
     print(list(squared_numbers))  
     o/p : [1, 4, 9, 16, 25]

    
     '''filter(): selects even numbers'''
     numbers = [1, 2, 3, 4, 5]
     even_numbers = filter(lambda x: x % 2 == 0, numbers)
     print(list(even_numbers))
     o/p : [2, 4]


     '''reduce(): calculates the product of all numbers'''
     from functools import reduce
     numbers = [1, 2, 3, 4, 5]
     product = reduce(lambda x, y: x * y, numbers)
     print(product)  
     o/p : 120

```python
# 4.Write a Python function that checks if a given number is prime or not from 1 to 200
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
for i in range(1,201):
  if is_prime(i):
    print(f"{i} is a prime number")
  else:
    print(f"{i} is not a prime number")
```

```python
'''
 5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.
'''
class FibonacciIterator:
    def __init__(self):
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        return value

fib_iter = FibonacciIterator()
n= int(input("Enter the required number of fibanocci series terms  "))
for _ in range(n):
    print(next(fib_iter))
```

```python
# 6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_generator(exponent):
    result = 1
    for i in range(exponent + 1):
        yield result
        result *= 2

exponent = int(input("Input the exponent value : "))

pow_gen = power_generator(exponent)

print(f"Powers of 2 up to exponent {exponent}:")
for power in pow_gen:
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.


def line_by_line_generator(text):
    # Split the text into lines
    lines = text.split('.')

    # Iterate through each line
    for line in lines:
      yield line

text = " Key sentences in academic writing do two things. First, they establish the main point that the rest of the paragraph supports. Second, they situate each paragraph within the sequence of the argument, a task that requires transitioning from the prior paragraph. Consider these two examples"
line_generator = line_by_line_generator(text)
for line in line_generator:
    print(line)
```
